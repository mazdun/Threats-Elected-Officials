```python
from pygooglenews import GoogleNews
import pandas as pd

gn = GoogleNews('en', 'US')
```

```python
#Assembling relevant search queries

#Notes: 

#I'm not including harassment as a keyword because this tends to return stories about officials resigning because
#of sexual harassment claims. I found that many of the articles that mention "harassment" also mention "threat."

search_queries = ['allintitle: resign threat', 'allintitle: resign threats', 
                  'allintitle: resigns threat', 'allintitle: resigns threats',
                  'allintitle: resigned threat', 'allintitle: resigned threats', 
                  'allintitle: resigning threat', 'allintitle: resigning threats', 
                  'allintitle: "step down" threat', 'allintitle: "step down" threats',
                  'allintitle: "steps down" threat', 'allintitle: "steps down" threats', 
                  'allintitle: "stepping down" threat', 'allintitle: "stepping down" threats', 
                  'allintitle: "stepped down" threat', 'allintitle: "stepped down" threats', 
                  'allintitle: quit threats', 'allintitle: quit threat', 
                  'allintitle: quits threat', 'allintitle: quits threats',
                  'intitle:mayor AND (intitle:resign OR intitle:resigns OR intitle:resigned OR intitle:resigning)',
                  'intitle:mayor AND (intitle:quit OR intitle:quits)',
                  'intitle:school AND intitle:board AND (intitle:resign OR intitle:resigns OR intitle:resigned OR intitle:resigning)',
                  'intitle:school AND intitle:board AND (intitle:quit OR intitle:quits)',
                  'intitle:sheriff AND (intitle:resign OR intitle:resigns OR intitle:resigned OR intitle:resigning)',
                  'intitle:sheriff AND (intitle:quit OR intitle:quits)',
                  'intitle:judge AND (intitle:resign OR intitle:resigns OR intitle:resigned OR intitle:resigning)',
                  'intitle:judge AND (intitle:quit OR intitle:quits)',
                  'intitle:commissioner AND (intitle:resign OR intitle:resigns OR intitle:resigned OR intitle:resigning)',
                  'intitle:commissioner AND (intitle:quit OR intitle:quits)',
                  'intitle:city AND intitle:council AND (intitle:resign OR intitle:resigns OR intitle:resigned OR intitle:resigning)',
                  'intitle:city AND intitle:council AND (intitle:quit OR intitle:quits)']
```

```python
#Looping over the search terms and retrieving the results

y = []

for x in search_queries:
    z=gn.search(x)
    b = z['entries']
    y.append(b)
    
```

```python
#converting to dataframe

df = pd.DataFrame()

for x in range(0,31):
    df = df.append(pd.DataFrame(y[x]), ignore_index=True)
    
    
```
